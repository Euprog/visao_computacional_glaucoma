```python
from skimage.io import imread,imsave, imshow
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
import os
from skimage import img_as_ubyte
```

```python
classe = 'glaucoma' #altere para a classe desejada

path = "/home/annyel/Área de Trabalho/Visão computacional/RIM-ONE_DL_images/partitioned_by_hospital/test_set/"+classe+'/'
names = os.listdir(path)

images = []

for name in names:
    images.append(imread(path+name))
```

```python
gray_images = []

for image in images:
    gray_images.append(rgb2gray(image))
```

```python
def potencia(imagem,gama,constante):
    return constante * imagem ** gama
```

```python
param = [(0.5, 1),(1.5,1),(2, 1),(2.5,1),(3,1),(4,1),(5,1)] # param = (gama)

for i in param:

    new_gray_images = []

    for image in gray_images: #potencia
        new_gray_images.append(potencia(image, i[0],i[1]))
    path = '/home/annyel/Área de Trabalho/Visão computacional/atividade/imagens_glaucoma_normal/trabalho_potencia/'+classe+str(' ')+str(i[0])+str(',')+str(i)+'/'
    os.mkdir(path)
    
    for j in range(len(new_gray_images)): #save images
        imsave(path+names[j], img_as_ubyte(new_gray_images[j]))
```
